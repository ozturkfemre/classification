# Train-Test Split

```python
from sklearn.model_selection import train_test_split

X = df.drop('class', axis=1) 
y = df['class'] 

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

# Classification Algorithms

## Classification Tree

```python
from sklearn.tree import DecisionTreeClassifier, plot_tree

treeclass = DecisionTreeClassifier(random_state=42)
treeclass.fit(X_train, y_train)
treeclass
```

```python
plt.figure(figsize=(35, 15))
plot_tree(treeclass, 
          feature_names=X_train.columns, 
          class_names=y_train.unique().astype(str),  # Dzeltilmi class_names
          filled=True, 
          rounded=True, 
          fontsize=9)
plt.show()
```

When the Classification Tree is analyzed:

-   The root node was identified as cp <= 0.5
-   Thalach being less than 168.5 was identified as one of the terminal nodes. However, it is noteworthy that there is no class change for both nodes. The situation is similar for the other terminal nodes. It is clear that the tree needs to be pruned.

-   age taking the value 42 was identified as one of the internal nodes.

```python
treeclass_pruned = DecisionTreeClassifier(
    random_state=42, 
    max_depth=5,  # Aa derinliini snrlama
    min_samples_leaf=10  # zellik saysn snrlama (gerekirse)
)

# Modeli eitme
treeclass_pruned.fit(X_train, y_train)

# Karar aacn grselletirme
plt.figure(figsize=(15, 10))
plot_tree(treeclass_pruned, 
          feature_names=X_train.columns, 
          class_names=y_train.unique().astype(str),  # Dzeltilmi class_names
          filled=True, 
          rounded=True, 
          fontsize=8)
plt.show()
```

When the classification tree after pruning is examined:

-   The root node was identified as cp <= 0.5
-   Thalach being less than 152.5 was identified as one of the terminal nodes. However, it is noteworthy that there is no class change for both nodes. The situation is similar for the other terminal nodes. It is clear that the tree needs to be pruned.

-   age <=63.5 was identified as one of the internal nodes.

```python
from sklearn.metrics import confusion_matrix, accuracy_score, f1_score
y_pred = treeclass_pruned.predict(X_test)
def calculate_metrics(y_pred, y_test):
    cm = confusion_matrix(y_test, y_pred)
    accuracy = accuracy_score(y_test, y_pred)
    # Sensitivity = TP / (TP + FN)
    sensitivity = cm[1, 1] / (cm[1, 1] + cm[1, 0])
    # Specificity = TN / (TN + FP)
    specificity = cm[0, 0] / (cm[0, 0] + cm[0, 1])
    precision = cm[1, 1] / (cm[1, 1] + cm[0, 1])
    f1 = f1_score(y_test, y_pred)
    classtree_results = [accuracy, sensitivity, specificity, precision, f1]
    return classtree_results
```

```python
classtree_results = calculate_metrics(y_pred, y_test)
classtree_results
```

## Bagging

```python
from sklearn.ensemble import BaggingClassifier
from sklearn.model_selection import cross_val_score
from sklearn.inspection import permutation_importance

bagging_model = BaggingClassifier(
    n_estimators=500,  
    oob_score=True, 
    random_state=42
)


bagging_model.fit(X_train, y_train)


oob_error = 1 - bagging_model.oob_score_


result = permutation_importance(bagging_model, X_train, y_train, n_repeats=10, random_state=42)


VI = pd.DataFrame({
    'var': X_train.columns,
    'imp': result.importances_mean
})

VI_plot = VI.sort_values(by='imp', ascending=False)


plt.figure(figsize=(10, 8))
plt.barh(VI_plot['var'], VI_plot['imp'], color='goldenrod')
plt.xlabel("Variable Importance")
plt.ylabel("Features")
plt.title("Variable Importance Plot")
plt.show()
```

```python
oob_error
```

- Differently from the classification tree, this time it is found that ca is the most important variable. 
- OOB error rate is found out as 0.24. for a dataset like tihs, it can be considered as low.

```python
y_pred = bagging_model.predict(X_test)
baggingresults= calculate_metrics(y_pred, y_test)
```

```python
baggingresults
```

## Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV

rf = RandomForestClassifier(n_estimators=500, max_features=4, random_state=42)

param_grid = {
    'max_features': [3, 4, 5, 6],
    'min_samples_split': [1, 3, 5, 10],
    'n_estimators': [200, 220, 300, 330]
}

grid_search = GridSearchCV(estimator=rf, param_grid=param_grid, scoring='neg_mean_squared_error', cv=10)
grid_search.fit(X_train, y_train)

best_params = grid_search.best_params_
print(f"Best Parameters: {best_params}")
print(f"Best RMSE: {-grid_search.best_score_}")

grid_results = pd.DataFrame(grid_search.cv_results_)
grid_results = grid_results[['param_max_features', 'param_min_samples_split', 'param_n_estimators', 'mean_test_score']]
grid_results['RMSE'] = np.sqrt(-grid_results['mean_test_score'])
print(grid_results.sort_values(by="RMSE").head(10))

# Best model based on grid search
best_rf = grid_search.best_estimator_

best_rf
```

```python
best_feature_importances = pd.DataFrame(best_rf.feature_importances_, index=X_train.columns, columns=["Importance"])
print(best_feature_importances)

best_feature_importances.sort_values(by="Importance", ascending=False).plot(kind='barh', color="goldenrod")
plt.title("Feature Importance for Best Model")
plt.xlabel("Importance")
plt.ylabel("Features")
plt.show()
```

When we examine the importance of the variables, we observe that ca and thal are the most important features just like the bagging. 

```python
y_pred = best_rf.predict(X_test)
rfresults = calculate_metrics(y_pred, y_test)
rfresults
```

## Support Vector Machines

```python
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV
# Linear Kernel
param_grid_linear = {
    'C': [0.1, 1, 10],
    'gamma': ['scale', 'auto'],
}

grid_search_linear = GridSearchCV(SVC(kernel='linear', probability=True), param_grid_linear, scoring='accuracy', cv=10)
grid_search_linear.fit(X_train, y_train)
best_linear = grid_search_linear.best_estimator_
```

```python
# Radial Kernel
param_grid_rbf = {
    'C': [0.1, 1, 10],
    'gamma': ['scale', 'auto'],
}

grid_search_rbf = GridSearchCV(SVC(kernel='rbf', probability=True), param_grid_rbf, scoring='accuracy', cv=10)
grid_search_rbf.fit(X_train, y_train)
best_radial = grid_search_rbf.best_estimator_
```

```python
y_pred = best_linear.predict(X_test)
linearresults = calculate_metrics(y_pred, y_test)
y_pred = best_radial.predict(X_test)
radialresults = calculate_metrics(y_pred, y_test)
```

```python
results = pd.DataFrame({
    'classification' : classtree_results,
    'bagging' : baggingresults,
    'random_forest' : rfresults,
    'svm:radial' : radialresults,
    'svm:linear' : linearresults,
     'xgb' : xgbresults
})
```

```python
results=results.T
results.columns = ['accuracy_score', 'sensitivity', 'specificity', 'precision', 'f1']
results
```

In the context of heart disease analysis, the best algorithm can be determined by examining multiple performance metrics, including accuracy, sensitivity (recall), specificity, precision, and F1 score. Based on the results, the random forest algorithm emerges as the best choice for the following reasons:

1. Accuracy:
		Random forest achieves the highest accuracy at 0.877778, indicating that it correctly classifies most of the instances compared to other models. This is crucial in medical diagnoses where overall correctness is important.

2. Sensitivity (Recall):
		Sensitivity measures the ability of the model to correctly identify true positive cases (patients with heart disease). Random forest has a high sensitivity of 0.829268, meaning it detects a large proportion of actual heart disease cases, only slightly lower than other models but balanced well with other metrics.

3. Specificity:
		Specificity refers to the ability to correctly classify healthy individuals (true negatives). Random forest excels here with a specificity of 0.918367, meaning it effectively avoids false positives (misclassifying healthy patients as sick). This is the highest among all models, making it highly reliable for ruling out heart disease in healthy individuals.

4. Precision:
		Precision indicates the proportion of true positives out of all positive predictions. Random forest has a high precision of 0.894737, meaning when it predicts heart disease, it is usually correct. This reduces the risk of false positives, which is important in preventing unnecessary treatments.

5. F1 Score:
		The F1 score balances sensitivity and precision. With an F1 score of 0.860759, random forest again stands out, showing that it strikes a good balance between identifying patients with heart disease and minimizing false positives.

Conclusion:

The random forest algorithm performs the best overall, as it maintains high accuracy, sensitivity, specificity, precision, and F1 score. It provides a strong balance between detecting actual heart disease cases and avoiding misclassifying healthy patients. This makes it highly suitable for heart disease diagnosis, where both false negatives (missing a diagnosis) and false positives (unnecessary treatment) can have serious consequences.

Comparison with Other Algorithms:
* SVM with a linear kernel and XGBoost are also strong performers but fall slightly behind random forest, especially in terms of F1 score and specificity.
* Bagging performs well but does not match random forest in terms of precision and F1 score.
* SVM with a radial kernel shows weaker sensitivity (0.536585), making it less effective at identifying heart disease cases.

Thus, random forest is the best algorithm for this heart disease classification task.
